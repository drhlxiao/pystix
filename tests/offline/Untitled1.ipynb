```python
import csv
fn='elut_run_219.csv'
offsets=[]
slopes=[]
with open(fn) as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        #print(row)
        print(row['offset'], row[' slope'])
        offsets.append(float(row['offset']))
        slopes.append(float(row[' slope']))
```

```python
from matplotlib import pyplot as plt
#print(slopes)
plt.plot(offsets)
plt.show()
plt.plot(slopes)
plt.show()
```

```python
import statistics as st
st.stdev([x for x in slopes if x<2.5])
```
